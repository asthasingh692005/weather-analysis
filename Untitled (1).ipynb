```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Load CSV File
# -----------------------------
df = pd.read_csv("weather.csv")

# Convert Date column to proper format
df['Date'] = pd.to_datetime(df['Date'])

# Show data
print("Weather Data:\n")
print(df)

# -----------------------------
# Step 2: Basic Analysis
# -----------------------------

# Average Temperature
avg_temp = np.mean(df['Temperature'])
print("\nAverage Temperature:", avg_temp)

# Maximum Temperature
max_temp = np.max(df['Temperature'])
print("Maximum Temperature:", max_temp)

# Minimum Temperature
min_temp = np.min(df['Temperature'])
print("Minimum Temperature:", min_temp)

# Hottest Day
hottest_day = df[df['Temperature'] == max_temp]
print("\nHottest Day:\n", hottest_day)

# Coldest Day
coldest_day = df[df['Temperature'] == min_temp]
print("\nColdest Day:\n", coldest_day)

# -----------------------------
# Step 3: Visualization
# -----------------------------

# 1. Temperature Trend (Line Graph)
plt.figure()
plt.plot(df['Date'], df['Temperature'])
plt.title("Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.xticks(rotation=45)
plt.show()

# 2. Humidity Trend (Line Graph)
plt.figure()
plt.plot(df['Date'], df['Humidity'])
plt.title("Humidity Trend")
plt.xlabel("Date")
plt.ylabel("Humidity")
plt.xticks(rotation=45)
plt.show()

# 3. Temperature Distribution (Histogram)
plt.figure()
plt.hist(df['Temperature'])
plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()

# 4. Wind Speed vs Temperature (Scatter Plot)
plt.figure()
plt.scatter(df['WindSpeed'], df['Temperature'])
plt.title("Wind Speed vs Temperature")
plt.xlabel("Wind Speed")
plt.ylabel("Temperature")
plt.show()
```
